# Data Cleaning

Data cleaning is an essential step in data processing.

The following operations are performed:
- Null value handling
- Duplicate removal
- Handling inconsistent data
- Data type verification

These steps improve data quality before analysis.

```python
from pyspark.sql.functions import col, sum

null_values = df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
])

null_values.show()
```

```python
print("Total Rows:", df.count())
print("Distinct Rows:", df.distinct().count())
```

```python
df = df.dropDuplicates()

print("Duplicate rows removed successfully")
```

```python
df = df.filter(col("Sales").isNotNull())

print("Null values handled successfully")
```

# Schema Modification

Schema modification is used to:
- Change column data types
- Rename columns
- Standardize dataset structure

This improves compatibility for analytical operations and aggregations.

```python
from pyspark.sql.types import DoubleType

df = df.withColumn(
    "Sales",
    col("Sales").cast(DoubleType())
)

df = df.withColumn(
    "Profit",
    col("Profit").cast(DoubleType())
)

print("Columns converted successfully")
```

```python
df = df.withColumnRenamed("Order ID", "Order_ID")

df = df.withColumnRenamed("Product Name", "Product_Name")

print("Columns renamed successfully")
```

```python
df.printSchema()
```

# Filtering Operations

Filtering is used to extract specific records from the dataset based on conditions.

PySpark filtering operations help in:
- Selecting important records
- Business-specific analysis
- Removing unnecessary data

```python
high_sales_df = df.filter(col("Sales") > 500)

high_sales_df.select(
    "Product_Name",
    "Category",
    "Sales",
    "Profit"
).show(5, truncate=False)
```

```python
furniture_df = df.filter(col("Category") == "Furniture")

furniture_df.show(5)
```

```python
loss_df = df.filter(col("Profit") < 0)

loss_df.select(
    "Product_Name",
    "Category",
    "Sales",
    "Profit"
).show(5, truncate=False)
```

```python
west_df = df.filter(col("Region") == "West")

west_df.show(5)
```

# Aggregations and groupBy Operations

Aggregation operations are used to summarize large datasets.

PySpark provides functions such as:
- sum()
- avg()
- count()
- min()
- max()

The groupBy() operation enables category-wise and region-wise analysis.

```python
region_sales = df.groupBy("Region").sum("Sales")

region_sales.show()
```

```python
avg_profit = df.groupBy("Category").avg("Profit")

avg_profit.show()
```

```python
segment_count = df.groupBy("Segment").count()

segment_count.show()
```

```python
from pyspark.sql.functions import max, min

df.select(
    max("Sales").alias("Maximum_Sales"),
    min("Sales").alias("Minimum_Sales")
).show()
```

```python
top_sales = df.orderBy(col("Sales").desc())

top_sales.select(
    "Product_Name",
    "Category",
    "Sales",
    "Profit"
).show(5, truncate=False)
```

# Wide Transformations

Wide transformations are Spark operations where data is shuffled between partitions.

These operations are computationally expensive because Spark redistributes data across executors before processing.

Examples of wide transformations:
- groupBy()
- join()
- distinct()

In this assignment, the `groupBy()` operation is a wide transformation because Spark performs shuffle operations before aggregation.

```python
wide_transformation = df.groupBy("Region").avg("Sales")

wide_transformation.show()
```

```python
distinct_categories = df.select("Category").distinct()

distinct_categories.show()
```

```python
print("Distinct Regions:", df.select("Region").distinct().count())
```

```python
# SQL Operations

PySpark allows SQL queries to be executed on DataFrames by creating temporary SQL views.

SQL operations are useful for:
- Analytical queries
- Business reporting
- Aggregation queries
- Structured data analysis

sales_query = spark.sql("""
SELECT Region,
       SUM(Sales) AS Total_Sales
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC
""")

sales_query.show()
```

```python
profit_query = spark.sql("""
SELECT Category,
       AVG(Profit) AS Average_Profit
FROM superstore
GROUP BY Category
ORDER BY Average_Profit DESC
""")

profit_query.show()
```

```python
top_products = spark.sql("""
SELECT Product_Name,
       Category,
       Sales,
       Profit
FROM superstore
ORDER BY Sales DESC
LIMIT 5
""")

top_products.show(truncate=False)
```

# Complete Data Processing Pipeline

The following pipeline combines:
- Data loading
- Data cleaning
- Schema modification
- Filtering
- Aggregation
- Sorting

into a unified Spark workflow.

This demonstrates an end-to-end data processing pipeline using PySpark.

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DoubleType

# Load Dataset
pipeline_df = spark.read.csv(
    "SampleSuperstore.csv",
    header=True,
    inferSchema=True,
    multiLine=True,
    escape='"'
)

# Remove Duplicate Rows
pipeline_df = pipeline_df.dropDuplicates()

# Handle Null Values
pipeline_df = pipeline_df.filter(col("Sales").isNotNull())

# Convert Columns to Numeric Types
pipeline_df = pipeline_df.withColumn(
    "Sales",
    col("Sales").cast(DoubleType())
)

pipeline_df = pipeline_df.withColumn(
    "Profit",
    col("Profit").cast(DoubleType())
)

# Rename Columns
pipeline_df = pipeline_df.withColumnRenamed(
    "Product Name",
    "Product_Name"
)

# Filter High Sales Records
pipeline_df = pipeline_df.filter(col("Sales") > 100)

# Aggregation
final_pipeline_result = pipeline_df.groupBy("Region") \
    .sum("Sales") \
    .orderBy("sum(Sales)", ascending=False)

# Display Final Result
final_pipeline_result.show()
```
